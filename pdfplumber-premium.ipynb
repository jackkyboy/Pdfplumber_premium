```python
# ฟังก์ชันเพื่ออ่านไฟล์ PDF และแปลงเป็นข้อความ
def read_pdf(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        # วนลูปผ่านแต่ละหน้าใน PDF
        for page in pdf.pages:
            # ดึงข้อความจากแต่ละหน้า
            text += page.extract_text() + "\n"
    return text

# ระบุเส้นทางไปยังไฟล์ PDF
file_path = "/Users/apichet/Downloads/อัตราเบี้ยซ่อมอู่ Pick up ,คอกเสมอหัวเก๋ง (รหัส 320) _(wef.01.04.2023).pdf"

# อ่านเนื้อหาใน PDF
document_text = read_pdf(file_path)

# แสดงข้อความที่อ่านได้จาก PDF ก่อนบันทึก
print("เนื้อหาของเอกสารมีดังนี้:\n")
print(document_text)

# นำข้อความที่อ่านได้ไปเก็บใน DataFrame ของ pandas
data = {'Document Text': [document_text]}
df = pd.DataFrame(data)

# แสดง DataFrame ก่อนบันทึก (เป็นตัวเลือกเพื่อดูผลลัพธ์)
print("\nDataFrame ก่อนบันทึก:\n")
print(df)

# หากต้องการบันทึก DataFrame ลงไฟล์ CSV
# df.to_csv("output.csv", index=False)
```

```python
def extract_information(text):
    data = {}

    # จับคู่ช่วงเวลาของ "ระยะเวลาแผนประกัน"
    match = re.search(r'ระยะเวลาแผนประกันตั้งแต่ (\d{1,2} \w+ \d{4})\s?-\s?(\d{1,2} \w+ \d{4})', text)
    if match:
        data['Insurance Period'] = f"{match.group(1)} - {match.group(2)}"
    
    # จับคู่ "ทุนประกันภัย" โดยกำหนดช่วงตัวเลขที่คาดหวัง
    match = re.search(r'ทุนประกันภัย\s*(\d{1,3},\d{3})\s*-\s*(\d{1,3},\d{3})', text)
    if match:
        data['Insurance Coverage'] = f"{match.group(1)} - {match.group(2)}"
    
    return data

# เรียกใช้ฟังก์ชันเพื่อแยกข้อมูล
extracted_info = extract_information(cleaned_text)

# แสดงข้อมูลที่แยกได้
print("ข้อมูลที่แยกได้:", extracted_info)
```
